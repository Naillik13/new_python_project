```python
usagersCaracteristiques2017 = pd.merge(left=caracteristiques2017 , 
         right=usagers2017, 
         left_on="Num_Acc",
         right_on="Num_Acc")
usagersCaracteristiques2017
```

```python
caracteristiques2017.isnull().sum().sum() # 23/ il y a 20911 valeurs manquantes  dans caracteristiques2017
```

```python
accicent_in_paris = usagersCaracteristiques2017.dep == 750
usagersCaracteristiques2017[accicent_in_paris]
```

```python
usagersCaracteristiques2017[accicent_in_paris].groupby(["grav"]).plot(kind="scatter", x="long", y="lat", alpha=0.2)
```

```python
beautiful_colors = [
    "#095231", "#a8d7b7",
    "#c0cefb", "#78b849",
    "#6bcd67", "#4b0502",
    "#7980e8", "#7da82f"]
```

```python
fig, ax = plt.subplots(figsize=(16,8), nrows=1, ncols=1)
for (name, group), color in zip(usagersCaracteristiques2017[accicent_in_paris].groupby(['grav']), beautiful_colors):
    # display on the same ax
    group.plot(
        kind='scatter',
        x='long',
        y='lat',
        label=name,
        color=color,
        ax=ax)
plt.legend(loc='best')
```
